```python
import pandas as pd

data = {
    'Year': [2020, 2020, 2021, 2021, 2022, 2023, 2023, 2024],
    'Startup': ['OpenAI', 'DeepMind', 'AI Labs', 'Vision AI', 'NeuralNet', 'Logic AI', 'AutoML Tech', 'Core AI'],
    'Funding_Round': ['Series A', 'Series B', 'Seed', 'Series A', 'Series B', 'Series C', 'Series B', 'Series C'],
    'Amount_Million_USD': [10.0, 15.0, 2.5, 8.0, 20.0, 35.0, 25.0, 40.0]
}

df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Aggregate total funding by year
funding_by_year = df.groupby('Year')['Amount_Million_USD'].sum().reset_index()

# Aggregate total funding by round
funding_by_round = df.groupby('Funding_Round')['Amount_Million_USD'].sum().reset_index()

# Order funding rounds
round_order = ['Seed', 'Series A', 'Series B', 'Series C', 'Series D']
funding_by_round['Funding_Round'] = pd.Categorical(funding_by_round['Funding_Round'], categories=round_order, ordered=True)
funding_by_round = funding_by_round.sort_values('Funding_Round')

# Plot total funding by year
plt.figure(figsize=(8,5))
sns.barplot(x='Year', y='Amount_Million_USD', data=funding_by_year, color='skyblue')
plt.title('Total AI Startup Funding by Year')
plt.ylabel('Funding Amount (Million USD)')
plt.xlabel('Year')
plt.show()

# Plot total funding by funding round
plt.figure(figsize=(8,5))
sns.barplot(x='Funding_Round', y='Amount_Million_USD', data=funding_by_round, color='salmon')
plt.title('Total AI Startup Funding by Funding Round')
plt.ylabel('Funding Amount (Million USD)')
plt.xlabel('Funding Round')
plt.show()
```
